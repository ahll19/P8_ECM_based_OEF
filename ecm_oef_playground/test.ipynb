```python
import numpy as np
import matplotlib.pyplot as plt
from pickle import load
```

```python
with open('fig/axes.pkl', 'rb') as f:
    axes = load(f)

with open('fig/axes_no-constraint.pkl', 'rb') as f:
    axes_no_constraint = load(f)
```

```python
fig, ax = plt.subplots(3, figsize=(10, 10), tight_layout=True)
for i in range(len(axes[1])):
    ax[0].plot(axes[1][i]-axes_no_constraint[1][i])

for i in range(len(axes[3])):
    ax[1].plot(axes[3][i]-axes_no_constraint[3][i])

for i in range(len(axes[5])):
    ax[2].plot(axes[5][i]-axes_no_constraint[5][i])
plt.show()
```
